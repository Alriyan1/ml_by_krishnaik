```python
df_copy=df.copy()
```

```python
df[df['Reviews']=='3.0M']
```

```python
df_copy.drop(df[df_copy['Reviews']=='3.0M'].index,inplace=True)
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'].value_counts()
```

```python
df_copy['Size']=df_copy["Size"].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df['Installs'].value_counts()
```

```python
df_copy['Price'].unique()
```

```python
char_to_remove=['+',',','$']
cols_in_clean=['Installs','Price']

for item in char_to_remove:
    for cols in cols_in_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy.info()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
#df_copy.to_csv('google_cleaned.csv')
```

```python
df_copy.head()
```

```python
df_copy[df_copy['App'].duplicated()]
```

```python
df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')
```

```python
df_copy.shape
```

```python
numeric_features=df_copy.select_dtypes(exclude='object').columns.to_list()
numeric_features
```

```python
categorical_features=df_copy.select_dtypes(include='object').columns.to_list()
categorical_features
```

```python
for col in categorical_features:
    print(col)
    print(df_copy[col].value_counts(normalize=True)*100)
    print("-------------------------")
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numerical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(0,len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_copy[numeric_features[i]],fill=True,color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of categorical Features',fontsize=20,fontweight='bold',alpha=0.8,y=1.)

category=['Type','Content Rating']
for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=df_copy[category[i]],palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
df_copy['Category'].value_counts().plot(kind='pie',figsize=(15,15),autopct='%0.1f%%')
```

```python
ax=df_copy['Category'].value_counts()[:10].plot(kind='bar',color=['r','b','g','y'])
for i in ax.containers:
    ax.bar_label(i)
```
