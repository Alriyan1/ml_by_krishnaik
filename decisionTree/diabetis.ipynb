```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
x=pd.DataFrame(dataset.data,columns=dataset['feature_names'])
```

```python
x
```

```python
y=dataset['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_trian,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_train.head()
```

```python
x_train.corr()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(x_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_trian)
```

```python
y_preds=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_preds)
```

```python
params={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(estimator=model,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(x_train,y_trian)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_preds=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,y_preds))
print(mean_absolute_error(y_test,y_preds))
print(mean_squared_error(y_test,y_preds))
```

```python
selectModel=DecisionTreeRegressor(criterion='squared_error',max_depth=5,max_features='log2',splitter='random')
```

```python
selectModel.fit(x_train,y_trian)
```

```python
plt.figure(figsize=(15,10))
from sklearn import tree
tree.plot_tree(selectModel,filled=True);
```
